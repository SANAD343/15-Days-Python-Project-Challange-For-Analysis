* What are all different subjects for which Udemy is offering courses ?

```python
# Install all required libraries and import the required data

import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Udemy Courses.csv')
```

```python
df.head(1)
```

# 1. What are all different subjects for which Udemy is offering courses?

```python
df.subject.unique()
```

# 2. Which subject has the maximum number of courses?

```python
df.subject.value_counts()
```

# 3. Show all the courses which are free of cost?

```python
df.head(2)
```

```python
df[df.is_paid == False]
```

```python
df[df.is_paid == False]
```

# 4. Show all the courses which are paid?

```python
df[df.is_paid == True]
```

# 5. Which are top selling courses?

```python
df.sort_values('num_subscribers', ascending = False)
```

# 6. Which are least selling courses?

```python
df.sort_values('num_subscribers', ascending = True)
```

# 7. Show all courses of graphic design where the price is below 100 

```python
df[(df.subject == 'Graphic Design') & (df.price < '100')]
```

# 8. Show all the courses of web development where the price is above 50?

```python
df[(df.subject == 'Web Development') & (df.price > '50')]
```

# 9. List out all the courses that are related with Python?

```python
len(df[df.course_title.str.contains('Python', case = True)])
```

# 10. What are courses that published in year 2015?

```python
df.dtypes
```

* We need to convert "published_timestamp" to datetime

```python
df['published_timestamp'] = pd.to_datetime(df.published_timestamp)  '''this function converts value of a column to datetime'''
```

```python
df.dtypes
```

```python
df['Year'] = df['published_timestamp'].dt.year
```

```python
df[df.Year == 2015]
```

* There are 1014 courses which are published in 2015

# 10. What are the maximum number of subscribers for each level of courses?

```python
df.level.unique()
```

```python
# df.groupby('level') - this is the syntax to create a group in any column.
# ['num_subscribers'] - this is the syntax to work on second column.

df.groupby('level')['num_subscribers'].max()
```

```python
df.groupby('level').max()
```
