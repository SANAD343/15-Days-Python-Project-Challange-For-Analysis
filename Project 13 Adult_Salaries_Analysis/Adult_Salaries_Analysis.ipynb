# Replace the Income Value ['<=50K', '>50k'] With 0 and 1 

```python
df['income'].value_counts()
```

```python
sns.countplot(x='income', data = df, palette='pastel', legend = False)
```

```python
# Method 1
```

```python
def income_data(inc):
    if inc == '<=50K':
        return 0
    else:
        return 1
```

```python
df['enconded_salary'] = df['income'].apply(income_data)
```

```python
df.head()
```

# Method 2 - Use replace method, 

* to_replace and values - to_replace=['<=50K','>50k']
* replace it with - value=[0,1]

```python
df.replace(to_replace = ['<=50K','>50k'], value = [0,1], inplace =True)
```

```python
df.head(1)
```

## Which Workclass Getting The Highest Salary?

```python
df.groupby('workclass')['enconded_salary'].mean().sort_values(ascending = False)
```

# Who Has Better Chance To Get Salary greater than 50K Male or Female?

```python
df.groupby('gender')['enconded_salary'].mean().sort_values(ascending = False)
```
