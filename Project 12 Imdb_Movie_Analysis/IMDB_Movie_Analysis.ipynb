## Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
df.columns
```

```python
# we are creating our own function to classify Movies based on Ratings.

# def rating - function name
# (rating) - let me pass argument

def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
df['rating_cat']=df['Rating'].apply(rating)

# you can use this created UDF using "apply" method
# lets create new column for this newly created category. "data['rating_cat']="
```

## Find Unique values from Genre

```python
df.Genre
```

Need to perform some steps to find unique values in Genre Column
* Split item with commas
* Convert this 2D list to 1D list to get unique values
* Find Unique values in empty list

```python
list1=[]
for value in df['Genre']: 
    list1.append(value.split(',')) # split the items by comms, then append the value over empty list
```

```python
list1
```

```python
# Convert 2D list to 1D list
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
# Find unique values in the empty list

uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```
